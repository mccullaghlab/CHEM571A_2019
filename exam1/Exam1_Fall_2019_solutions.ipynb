**b.** Determine the commutator of the following two operators
    
**i.** $\hat{A} = \frac{1}{x} $, $\hat{B} = \frac{d^2}{dx^2} $

We start by recognizing that the commutator of $\hat{A}$ and $\hat{B}$ is $[\hat{A},\hat{B}] = \hat{A}\hat{B} - \hat{B}\hat{A}$.  In order to determine the single operator that represents this commutator, we must apply the commutator to a function, $f$, of $x$.

\begin{align}
[\hat{A},\hat{B}]f &=& \left(\hat{A}\hat{B} - \hat{B}\hat{A}\right)f\\
 &=& \left( \frac{1}{x}\frac{d^2}{dx^2} - \frac{d^2}{dx^2}\frac{1}{x}\right)f\\
&=& \frac{1}{x}\frac{d^2f}{dx^2} - \frac{d^2}{dx^2}\left(\frac{f}{x}\right)\\
&=& \frac{1}{x}\frac{d^2f}{dx^2} - \frac{d}{dx}\left( -\frac{f}{x^2} + \frac{1}{x}\frac{df}{dx}\right)\\
&=& \frac{1}{x}\frac{d^2f}{dx^2} - \left(\frac{2f}{x^3} - \frac{1}{x^2}\frac{df}{dx} - \frac{1}{x^2}\frac{df}{dx} + \frac{1}{x}\frac{d^2f}{dx^2}\right) \\
&=& -\frac{2f}{x^3} + \frac{2}{x^2}\frac{df}{dx}\\
&=& \left(-\frac{2}{x^3} + \frac{2}{x^2}\frac{d}{dx}\right)f
\end{align}

$\Rightarrow [\hat{A},\hat{B}] = -\frac{2}{x^3} + \frac{2}{x^2}\frac{d}{dx}$
        
**ii.** $\hat{A} = 2x^2 $, $\hat{B} = (1+x)\frac{d}{dx} $

Follow the same procedure as above.

\begin{align}
[\hat{A},\hat{B}]f &=& \left(2x^2(1+x)\frac{d}{dx} - (1+x)\frac{d}{dx}2x^2\right) f \\
&=& 2x^2(1+x)\frac{df}{dx} - (1+x)\frac{d}{dx}\left( 2x^2f \right) \\
&=& 2x^2(1+x)\frac{df}{dx} - (1+x)\left( 4xf + 2x^2\frac{df}{dx}\right) \\
&=& - 4x(1+x)f
\end{align}

$\Rightarrow [\hat{A},\hat{B}] = - 4x(1+x)$

   **c.** Develop a set of basis functions over the domain $0<r<10$ that can be used to describe $V(r)$.  Write out the basis set and plot it over domain $1<r<10$.
   
   Note that the wording of this question is misleading (not on purpose).  We want basis functions that describe the wavefunction, not the potential.  So what happens to the wavefunction at the boundaries?  Well, as $V(r)$ approaches infinity, the wavefunction should approach zero and similarly as the potential approaches zero the wavefunction should also be zero.  Thus we need basis functions, $g_i(r)$, that have the following properties:
   
$ \lim_{r\rightarrow0}g_i(r)= 0 \quad \forall i=0,1,..,N$

$ \lim_{r\rightarrow\infty}g_i(r)= 0 \quad \forall i=0,1,..,N$

There are a few options to choose from but I choose to use

$g_i(r) = N_i r\exp(-\alpha (r-r_i)^2)$,

where $N_i$ is the normalization constant and $r_i$ shifts the position of the gaussian component of the basis functions.  Note that to completely define my basis set I need to choose $alpha$, $N$, the total number of basis functions, as well as the positions of each basis function.  I will opt to keep the domain ($0<r<10$) over which my basis functions span constant vary the spacing when I increase the number of basis functions to check convergence.

```python
# define basis functions
alpha = 1.0 # arbitrary choice
def basis(r,ri,alpha):
    return r*np.exp(-alpha*(r-ri)**2)
# basis function squared for normalization purposes
def basis2(r,ri,alpha):
    return r*r*np.exp(-2*alpha*(r-ri)**2)
```

```python
# plot ten basis functions 
from scipy import integrate
ax =  define_figure(xlabel="r (Bohr)",ylabel="g(r)")
dr = 10.0 / 10
r = np.arange(1,10,0.01)
for i in range(10):
    ri = i*dr
    norm = np.sqrt(integrate.quad(basis2,0,np.inf,args=(ri,alpha))[0])
    label = "basis function " + str(i+1)
    plt.plot(r,basis(r,ri,alpha)/norm,lw=2,label=label)
plt.legend(fontsize=16)
```

**d.** Determine the variational $S_{ij}$ and $H_{ij}$ for this basis set and Hamiltonian.

This is completely analytical work until a point.  We must derive the two matrix elements in our basis.  We start with $\mathbf{S}$.

\begin{align}
S_{ij} &=& \langle g_i | g_j \rangle \\
&=& \int_0^\infty N_i r\exp(-\alpha (r-r_i)^2)N_j r\exp(-\alpha (r-r_j)^2)dr \\
&=& N_iN_j\int_0^\infty r^2\exp(-\alpha (r-r_i)^2)\exp(-\alpha (r-r_j)^2)dr \\
&=& N_iN_j\exp\left(\frac{-\alpha(r_i-r_j)^2}{2}\right)\int_0^\infty r^2\exp\left(2\alpha\left(r-\frac{x_i+x_j}{2}\right)^2\right)dr
\end{align}

Where the last step is from recognizing the product of two gaussians is a gaussian centered at the midpoint between the two original gaussians.  While this may have a closed form, we will use numerical integration to solve this integral.  Let's demonstrate in code below that it is indeed closed.

```python
def S_integrand(r,ri,rj,alpha):
    return basis(r,ri,alpha)*basis(r,rj,alpha)

alpha = 10
r0 = 0.0
r1 = 1.0
S01 = integrate.quad(S_integrand,0,np.inf,args=(r0,r1,alpha))[0]
print("S01=", S01)
```

We now work on $\mathbf{H}$.

\begin{align}
H_{ij} &=& \langle g_i |\hat{H}| g_j \rangle \\
&=& \langle g_i |\frac{\hat{p}^2}{2\mu} + \frac{q_1q_2}{r} + \frac{\sigma}{r^6}| g_j \rangle \\
&=& \langle g_i |\frac{\hat{p}^2}{2\mu}|g_j\rangle + \langle g_i |\frac{q_1q_2}{r}|g_j\rangle + \langle g_i |\frac{1}{r^6}|g_j\rangle \\
&=& K_{ij} + V^C_{ij} + V^{ex}_{ij}
\end{align}

Where the last step is simply defining three different integrals that comprise the Hamiltonian matrix element.  These are the kinetic energy integral, $K_{ij}$, the Coulomb integral, $V^C_{ij}$, and the excluded volume integral, $V^{ex}_{ij}$. The latter two are easily solved using numerical integration so we will not concern ourselves with those.  The kinetic energy matrix element, however, requires us to perform a second derviative.

\begin{align}
K_{ij} &=& \langle g_i |\frac{\hat{p}^2}{2\mu}|g_j\rangle \\
&=& \frac{-\hbar^2}{2\mu}\langle g_i |\frac{d^2}{dr^2}|g_j\rangle \\
\end{align}

Let's condiser the second derivative

\begin{align}
\frac{d^2}{dr^2}g_j &=&  \frac{d^2}{dr^2} \left(N_jr\exp\left(-\alpha(r-r_j)^2 \right)\right) \\
&=& N_j \frac{d}{dr} \left(\exp\left(-\alpha(r-r_j)^2\right) -2\alpha r(r-r_j)\exp\left(-\alpha(r-r_j)^2\right)\right) \\
&=& N_j \frac{d}{dr} \left(  \exp\left(-\alpha(r-r_j)^2\right) -2\alpha r^2\exp\left(-\alpha(r-r_j)^2\right) + 2\alpha rr_j\exp\left(-\alpha(r-r_j)^2\right) \right) \\
&=& N_j \exp\left(-\alpha(r-r_j)^2\right)\left( -2\alpha(r-r_j) -4\alpha r  +4\alpha^2r^2(r-r_j) + 2\alpha r_j -4\alpha^2rr_j(r-r_j)\right) \\
&=& 2N_j\alpha\exp\left(-\alpha(r-r_j)^2\right)\left(-(r-r_j) - 2r + 2\alpha r^2(r-r_j) + r_j - 2\alpha rr_j(r-r_j)\right) \\
&=& 2N_j\alpha\exp\left(-\alpha(r-r_j)^2\right)\left(-3r + 2r_j + 2\alpha r(r-r_j)^2\right) \\
\end{align}

Thus we should have

$ K_{ij} = \frac{-\hbar^2}{2\mu} \int_0^\infty g_i g_j 2\alpha \left(-3 + 2\alpha (r-r_j)^2\right)dr$

$ V^C_{ij} = q_1q_2\int_0^\infty g_i g_j r^{-1}dr$

$ V^{ex}_{ij} = \sigma\int_0^\infty g_i g_j r^{-6}dr$

```python
# code for Hamiltonian components
def K_integrand(r,ri,rj,alpha):
    return basis(r,ri,alpha)*2*alpha*(np.exp(-((r-rj)**2)))*(-3*r + 2*rj + 2*alpha*r*(r-rj)**2)
def VC_integrand(r,ri,rj,alpha):
    return basis(r,ri,alpha)*basis(r,rj,alpha)*r**(-1)
def VEX_integrand(r,ri,rj,alpha):
    return basis(r,ri,alpha)*basis(r,rj,alpha)*r**(-6)
def V_integrand(r,ri,rj,alpha):
    return basis(r,ri,alpha)*basis(r,rj,alpha)*(r**(-6)-r**(-1))
```

```python
# check to see that our matrices are Hermitian
nBasis=4
alpha = 1.0
hbar = 1.0
massProton = 1.0/5.45E-4
mu = massProton/2.0
K = np.empty((nBasis,nBasis))
V = np.empty((nBasis,nBasis))
S = np.empty((nBasis,nBasis))
dr = 10.0/nBasis
for i in range(nBasis):
    ri = i*dr
    for j in range(nBasis):
        rj = j*dr
        K[i,j] = -0.5/mu*integrate.quad(K_integrand,0,np.inf,args=(ri,rj,alpha))[0]
        V[i,j] = integrate.quad(V_integrand,0,np.inf,args=(ri,rj,alpha))[0]
        S[i,j] = integrate.quad(S_integrand,0,np.inf,args=(ri,rj,alpha))[0]
print(K-K.T)
print(V-V.T)
print(S-S.T)
```

**e.** Solve the variational problem.  Compute g.s. energy and plot wavefunction.

This means to populate the $\mathbf{H}$ and $\mathbf{S}$ matrices and diagonalize $\mathbf{S}^{-1}\mathbf{H}$.

```python
# variational principle basis set solution 
hbar = 1.0
massProton = 1.0/5.45E-4
mu = massProton/2.0
def basis_variational_simps(nBasis,alpha = 1.0,minR=1E-2,maxR=1E5,deltaR=1E-2):
    K = np.empty((nBasis,nBasis))
    V = np.empty((nBasis,nBasis))
    S = np.empty((nBasis,nBasis))
    dr = 10.0/nBasis
    rIntegrand = np.arange(minR,maxR,deltaR)
    for i in range(nBasis):
        ri = (i+0.5)*dr
        for j in range(i,nBasis):
            rj = (j+0.5)*dr
            K[i,j] = -0.5/mu*integrate.simps(K_integrand(rIntegrand,ri,rj,alpha),rIntegrand)
            K[j,i] = K[i,j]
            V[i,j] = integrate.simps(V_integrand(rIntegrand,ri,rj,alpha),rIntegrand)
            V[j,i] = V[i,j]
            S[i,j] = integrate.simps(S_integrand(rIntegrand,ri,rj,alpha),rIntegrand)
            S[j,i] = S[i,j]
    H = K + V
    SinvH = np.dot(np.linalg.inv(S),H)
    # compute eigenvalues and eigenvectors
    H_eig_val, H_eig_vec = np.linalg.eig(SinvH)
    # reorder these so largest eigenvalue is first
    idr = H_eig_val.argsort()   
    H_eig_val = H_eig_val[idr]
    H_eig_vec = H_eig_vec[:,idr]
    return H_eig_vec, H_eig_val, S
```

Above is a subroutine to perform the variational approach for a specified number of basis functions.  In order to suggest that we have a reasonable ground state energy we must check convergence as a function of number of basis functions.  I will run 4 to 19 basis functions and plot the g.s. energy.

```python
energyConvergence = []
for i in range(4,20):
    psi, E, S = basis_variational_simps(i,5.0)
    energyConvergence.append(E[0])
    print("Ground state energy for i=", i, " is:", E[0])
```

```python
ax = define_figure(xlabel="Number of Basis Functions",ylabel="Energy (a.u)")
ax.plot(np.arange(4,20,1),energyConvergence,'-o')
```

```python
energyConvergence = []
for i in range(4,28):
    psi, E, S = basis_variational_simps(i,5.0,minR=1E-4,maxR=1E4,deltaR=1E-1)
    energyConvergence.append(E[0])
    print("Ground state energy for i=", i, " is:", E[0])
```

```python
ax = define_figure(xlabel="Number of Basis Functions",ylabel="Energy (a.u)")
ax.plot(np.arange(4,28,1),energyConvergence,'-o')
```

   **f.** Determine $\langle r^2 \rangle$ for the ground state and first excited state of this system.
   
   For this we simply need to recognize:
   
   $ \langle r^2 \rangle^{(n)} = \int_0^\infty \psi^{(n)}(r)r^2\psi^{(n)}(r)dr $
   
   And then solve this using our approximate variational wavefunctions.

```python
r = np.arange(1E-6,1E3,1E-3)
for n in range(0,2):
    # normalize the wavefunctions
    norm = 0.0
    for i in range(nBasis):
        for j in range(nBasis):
            norm += psi[i,n]*psi[j,n]*S[i,j]
    norm = np.sqrt(norm)
    # compute wavefunction as function of r
    psiR = np.zeros(r.size)
    dr = 10.0/nBasis
    for i in range(nBasis):
        ri = (i+0.5)*dr
        psiR += psi[i,n]*basis(r,ri,alpha)/np.sqrt(S[i,i])
    r2 = integrate.simps(psiR*np.power(r,2)*psiR,r)
    print(r2)
    
```
